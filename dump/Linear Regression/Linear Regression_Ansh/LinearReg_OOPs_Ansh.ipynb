```python
#Training a linear regression model using scikit learn

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import preprocessing
```

```python
#importing csv file

df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Real estate.csv")
```

```python


print(df.head())
print(df.columns)
```

```python
#plots of data
sns.scatterplot(x='X4 number of convenience stores', y='Y house price of unit area', data=df)
```

```python
# for feature variable

X= df.drop('Y house price of unit area',axis=1)
Y= df['Y house price of unit area']

print(X)

print(Y)
```

```python
#splitting the data into training and test sets for better evaluation

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=101)

```

```python
#Creating a LinearRegression model

model = LinearRegression()
```

```python
#fitting the model
model.fit(X_train,y_train)
```

```python
#making prediction on test data

y_pred = model.predict(X_test)
```

```python
#evaluation of error
from sklearn.metrics import mean_squared_error, mean_absolute_error

print('mean_squared_error : ', mean_squared_error(y_test, y_pred))

print('mean_absolute_error : ', mean_absolute_error(y_test, y_pred))
```

```python
plt.scatter(y_test,y_pred)
plt.xlabel('Actual')
plt.ylabel('Predicted')
```

```python
#Regression plot
sns.regplot(x=y_test,y=y_pred,ci=None,color ='red')
```
